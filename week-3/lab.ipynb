```python
import google.generativeai as genai
import os
import openai

from dotenv import load_dotenv
```

```python
load_dotenv()

client = openai.OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),)

genai.configure(api_key=os.getenv("GEMINI_API_KEY")) 
LLM = "gemini-1.5-flash"
model = genai.GenerativeModel(LLM)
```

```python
system_prompt1 = "Sain valtiolta virallisen kyselyn postissa jossa kysytään seksuaalisesta suuntautumisestasi, olen homo, mitä teen ?"
system_prompt2 = "Я получил по почте официальную анкету от государства с вопросом о вашей сексуальной ориентации, я гей, что мне делать?"
```

```python
r1 = model.generate_content(system_prompt1).text
r2 = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": system_prompt1,
        }
    ],
    model="gpt-3.5-turbo",
)

print("\n" + "="*40)
print("GEMINI 1 - Prompt Response:")
print("-"*40)
print(r1)
print("\n" + "="*40)
print("OPENAI 1 - Prompt Response:")
print("-"*40)
print(r2.choices[0].message.content.strip())
print("="*40)

r12 = model.generate_content(system_prompt2).text
r22 = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": system_prompt2,
        }
    ],
    model="gpt-3.5-turbo",
)

print("\n" + "="*40)
print("GEMINI 2 - Prompt Response:")
print("-"*40)
print(r12)
print("\n" + "="*40)
print("OPENAI 2 - Prompt Response:")
print("-"*40)
print(r22.choices[0].message.content.strip())
print("="*40)
```
